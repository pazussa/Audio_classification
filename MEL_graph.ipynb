```python
import numpy as np
import librosa
import IPython.display as ipd

%matplotlib inline
import matplotlib.pyplot as plt
import librosa.display
```

```python
import pandas as pd
train_features = pd.read_csv('/content/drive/MyDrive/Uninorte/Todo_cuenta_uninorte/UrbanSound8K/metadata/UrbanSound8K.csv') # Cargar archivo CSV como DataFrame pandas (matriz)
classes = train_features.groupby(['class']) # Agrupar clases en un grupo
for class_name in classes.groups:   # Iterar sobre cada clase
    print('Clase : ', class_name)     # Mostrar nombre de la clase actual (iteración)
    instance = classes.get_group(class_name)['slice_file_name'].tolist()[0]  # Tomar primer archivo de audio perteneciente a esta clase
    print('Ejemplo : ', instance)     # Mostrar nombre del archivo (audio) actual (iteración)
 
```

```python

import matplotlib.pyplot as plt
from IPython.display import Audio
%matplotlib inline 
# Cargar audio de ejemplo: air_conditioner
x, sr = librosa.load('/content/drive/MyDrive/Uninorte/Todo_cuenta_uninorte/UrbanSound8K/audio/fold5/100852-0-0-0.wav')   # Or your local file path
n_fft=2048; hop_length=112; n_mels=128; duration = 2 # Segundos a graficar en el espectrograma (2 segundos)
mel_spec = librosa.feature.melspectrogram(x, sr=sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels)
mel_spec = mel_spec[:,:int(duration*sr/hop_length)] # Tomar los primeros 2 segundos del espectrograma
mel_spec = librosa.power_to_db(mel_spec, ref=np.max)  # Convertir a dB
plt.figure(figsize=(5, 4))                            # Tamaño de la figura en pulgadas
librosa.display.specshow(mel_spec, sr=sr, hop_length=hop_length, x_axis='time', y_axis='mel') # Graficar espectrograma
plt.colorbar()                                        # Mostrar barra de colores en la figura
plt.show()                                            # Mostrar figura
```
